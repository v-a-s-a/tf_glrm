# PCA

We can fit a PCA model in the GLRM framework. We'll test on the well known MNIST conveniently provided in the `tensorflow.examples.tutorials` module.

The variable notation is chosen to match the original GLRM paper.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)


N = 784      # features
M = 100      # batch size
K = 2        # latent dimension
steps = 1000 # gradient descent steps


g = tf.Graph()
with g.as_default():

    # observed representation
    A = tf.placeholder(dtype=tf.float64, shape=(M, N))
    
    # archetypical representations
    X = tf.get_variable("X", dtype=tf.float64, shape=(M, K))
    Y = tf.get_variable("Y", dtype=tf.float64, shape=(K, N))
    
    # data reconstruction
    Z = tf.matmul(X, Y)
    
    # objective
    loss = tf.reduce_mean((Z - A)**2)
    grad_op = tf.train.GradientDescentOptimizer(0.01).minimize(loss)
    

with tf.Session(graph=g) as sess:
    # setup
    sess.run(tf.global_variables_initializer())
    losses = np.zeros(steps)
    
    # estimation
    for step in range(steps):
        batch_xs, batch_ys = mnist.train.next_batch(M)
        _, losses[step] = sess.run([grad_op, loss], feed_dict={A: batch_xs})

        
# summarize training
plt.figure()
plt.scatter(x=range(steps), y=losses)
plt.show()
```
